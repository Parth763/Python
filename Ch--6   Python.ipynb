```python
#   '#' new file ma na aavva joy
#     f.seek(0) :- Pointer ne first leter pr lai ne jay.

f=open('Me.txt','r')
f1=open('Ex1.txt','w+')

for i in f:
    n=i.find('#')
    if n==-1:
        f1.write(i)
    elif i[0]=='#':
        f1.write("")
    else:
        f1.write(i[:n]+'\n')
        
f.seek(0)
f1.seek(0)

print(f.read())
print("------New File------")
print(f1.read())

f.close()
f1.close()
```
